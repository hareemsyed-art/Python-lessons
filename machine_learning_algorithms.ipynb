linear regression

```python
import numpy as np
from sklearn.linear_model import LinearRegression
age = np.array([[6], [7], [8], [9], [10]])
height = np.array([1000, 1500, 2000, 2500, 3000])
model = LinearRegression()
model.fit(age, height)
print("height if person was 11", model.predict([[11]]))
```

K nearest neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
x = np.array([
    [200, 5],
    [250, 4],
    [150, 6],

    [50, 8],
    [40, 7],
    [60, 9]
])
y = np.array([0, 0, 0, 1, 1, 1])
model = KNeighborsClassifier(n_neighbors = 4)
model.fit(x, y)
new_fruit = [[160, 5]]
pred = model.predict(new_fruit)
print(pred)
for i in pred:
  if i == 0:
    print("Watermelon")
  else:
    print("Mango")
```
