```python

import matplotlib.pyplot as plt
import pandas as pd
data = pd.DataFrame({
    "movie" : ["Avengers", "Frozen", "Batman", "Toy Story", "Spider-Man", "Inception"],
    "genre" : ["Action", "Animation", "Action", "Animation", "Action", "Sci-Fi"],
    "budget" : [220, 150, 200, 120, 180, 160],
    "ratings" : [8.4, 7.5, 8.2, 8.3, 8.1, 8.8],
    "earned" : [2797, 1450, 1005, 1067, 1900, 836]
})
print(data)
a = pd.pivot_table(data, values = ["budget", "ratings", "earned"], index = "genre", aggfunc = "mean")
print(a)
a.plot(kind = "bar", figsize = (8,8))
plt.title("avg budget, rating, and earned for action, animation, sci-fi")
plt.ylabel("mean of budget, earned, and ratings")
plt.show()

b = pd.pivot_table(data, values = "ratings", index = "genre", aggfunc = "sum")
print(b)
b.plot(kind = "bar", figsize = (9,9))
plt.title("sum of Animation, Action, Sci-Fi")
plt.ylabel("sum of animation, Action, Sci-Fi")
plt.xticks(rotation = 45)
plt.show()
```
