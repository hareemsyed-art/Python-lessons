```python
#       Part 1 – Practice Dataset
# =====================================

# Create a pandas DataFrame using this data.
import pandas as pd
a = pd.DataFrame({
    "city" : ["Karachi", "Lahore", "Islamabad", "Dubai", "London"],
    "temperature" : [34, 32, 30, 38, 22],
    "rainfall" : [12, 20, 30, 5, 50],
    "population_millions" : [16, 13, 1.2, 3.5, 9],
    "air_quality_index" : [110, 160, 95, 80, 60]
})
print(a)
```

```python
# Make These Graphs
import seaborn as sns
import matplotlib.pyplot as plt

# plt.subplot(fig, axes)
fig, axes = plt.subplots(2, 2, figsize = (10, 10))

# 1. Histogram
# Goal: See how temperatures are distributed.
# Axes help
# X-axis → temperature
# Y-axis → count of cities
sns.histplot(a["temperature"], ax = axes[0, 0])
axes[0,0].set_title("temperature")

# 2. Scatter Plot
# Goal: See if higher temperatures relate to air quality.
# Axes help
# X-axis → temperature
# Y-axis → air_quality_index
sns.scatterplot(x = "temperature", y = "air_quality_index", data = a, ax = axes[0, 1])
axes[0, 1].set_title("temperature and air quality index")

# 3. Bar Chart
# Goal: Compare population of each city.
# Axes help
# X-axis → city
# Y-axis → population_millions
a["city"].value_counts().plot(kind = "bar", ax  = axes[1, 0])
axes[1, 0].set_title("Compared population of each city")



# 4. Box Plot
# Goal: Compare air quality across cities.
# Axes help
# X-axis → city
# Y-axis → air_quality_index
sns.boxplot(x = "city", y = "air_quality_index", data = a, ax = axes[1, 1])
axes[1, 1].set_title("air quality across cities")
```
